# Cumulative Miles

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a pandas Series to store cumulative miles
miles = pd.Series([55, 120, 200, 280, 375, 475, 600]) 
```

```python
# Calculate the daily miles by taking the difference between elements
daily_miles = miles.diff()
```

```python
# Set first entry to NaN and fill it with the first value from miles
daily_miles[0] = pd.NA
daily_miles.iloc[0] = miles.iloc[0]

print("Cumulative Miles:")
print(miles)

print("\nDaily Miles:") 
print(daily_miles)
```

## Line Plot for Cumulative Miles

```python
# Line plot of cumulative miles
plt.subplot(1,2,1)
miles.plot()
plt.title("Cumulative Miles")
plt.ylabel("Miles")

plt.tight_layout()
plt.show()
```

## Bar Plot for Daily Miles

```python
# Bar chart of daily miles
plt.subplot(1,2,2)
daily_miles.plot.bar()
plt.title("Daily Miles")
plt.ylabel("Miles")

plt.tight_layout()
plt.show()
```
